```python
#df=df.drop(['Countries and Areas','Value','Under5 _M_Rank'], axis=1)
df=df[['Under5_MR_2016_per1000','Under5 _M_Rank']]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#data_scaled = scaler.fit_transform(df)
data_scaled=df
```

```python
pd.DataFrame(df).describe()
```

```python
from sklearn.cluster import KMeans

# fitting multiple k-means algorithms and storing the values in an empty list
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 3, init='k-means++')
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)

frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()
```

```python
frame['Countries and Areas']=data['Countries and Areas']
data['cluster'] = pred
Visuals['cluster'] = pred
Visuals.to_csv('Visuals_3.csv', index=False)
```
